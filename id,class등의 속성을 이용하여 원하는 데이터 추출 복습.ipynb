```python
import requests
from bs4 import BeautifulSoup 
```

```python
url ='https://sports.news.naver.com/news.nhn?oid=436&aid=0000034943'

resp=requests.get(url)
resp.text
```

```python
url ='https://sports.news.naver.com/news.nhn?oid=436&aid=0000034943'
resp=requests.get(url)

soup=BeautifulSoup(resp.text)

container=soup.find('h4',class_='title')
container.get_text()
```
